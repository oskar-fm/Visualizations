```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simple array
x = np.array([1, 2, 3, 4])
```

```python
#Sample #2
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])
```
